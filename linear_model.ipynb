### Linear model using the continous variables from the data

```python
import  pandas  as  pd
import  numpy  as  np
```

```python
df = pd.read_csv("count_data.csv")
df = df[["points","price","word_count","age"]]
df.head()
```

```python
X = df[['age', 'price']]
y = df['points']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# from sklearn.preprocessing import PowerTransformer
# pt = PowerTransformer()
# X_trans = pt.fit(X_train).transform(X_test)
```

```python
from sklearn import linear_model
from sklearn.metrics import r2_score
reg = linear_model.Ridge()
model = reg.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(f'R-Squared for test data: {model.score(X_test, y_test)}')
print(f'R-Squared for predictions: {r2_score(y_test, y_pred)}')
```

```python
X_test["pred"] = pd.Series(model.predict(X_test))
X_test["score"] = pd.Series(y_test)
results_df = X_test[["pred","score"]]
results_df.dropna(inplace=True)
results_df.head(20)
```
