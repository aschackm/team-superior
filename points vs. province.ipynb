```python
import  pandas  as  pd
import  numpy  as  np
import  os
%matplotlib  inline
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
import pyLDAvis.sklearn
```

```python
df = pd.read_csv(os.path.join("wineData.csv"))
df.head()
```

```python
df.count()
```

```python
new = df.dropna()
```

```python
new.count()
```

```python
province_df = new[["points", "province"]]
province_df.head()
```

```python
province_df.count()
```

```python
province_df.sort_values(by='points', ascending=False)
```

```python
grouped = province_df.groupby('province').mean()
grouped
```

```python
top_province = grouped.sort_values(by='points', ascending=False)
top_province.head(10)
```

```python
variety_df = new[["points", "variety"]]
variety_df.head()
```

```python
group_variety = variety_df.groupby('variety').mean()
group_variety.head(10)
```

```python
top_variety = group_variety.sort_values(by='points', ascending=False)
top_variety.head(10)
```

```python
winery_df = new[["points", "winery", "country"]]
winery_df.head()
```

```python
group_winery = winery_df.groupby('winery').mean()
group_winery.head()
```

```python
top_winery = group_winery.sort_values(by='points', ascending=False)
top_winery.head(10)
```
